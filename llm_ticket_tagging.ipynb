## Import Library

```python
from transformers import pipeline
```

## Load Zero-shot classifier

```python
classifier = pipeline('zero-shot-classification')
```

```python
# defibbe Labels
labels = [
    'Billing Issue',
    'Technical Issue',
    'Account Management',
    'Complaint',
    'Network Problem',
    'Payment Issue'
]
```

```python
def classify_ticket(text):
    result = classifier(text, labels)

    # Top 3 tags
    top_labels = result['labels'][:3]
    scores = result['scores'][:3]

    print('\nTicket:', text)
    print('Top 3 Tags:')
    
    for label, score in zip(top_labels, scores):
        print(f'{label} ({round(score,2)})')
```

## Test

```python
tickets = [
    'My internet is not working since yesterday',
    'I was charged twice for my bill',
    'Unable to login to my account'
]

for t in tickets:
    classify_ticket(t)
```
